## Graphische Aufarbeitung von Daten

```python
sns.pairplot(df,vars=["price", "number_of_reviews", "availability_365"])
#Es werden hier die drei Variablen jeweils gegeneinander geplottet
#wenn eine Variable gegen eine andere geplottet wird, dann wird ein "Scatterplot" erzeugt
#Wenn die Variable gegen sich selber geplottet wird, dann wird ein Histogramm erzeugt
```

```python
#Erzeugen einer "Heat-Map"
df_corr = df.corr()
sns.heatmap(df_corr,annot=True,cmap="coolwarm")
```
